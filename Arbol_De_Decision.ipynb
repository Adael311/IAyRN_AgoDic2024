HACEMOS LA IMPORTACION DE LIBRERIAS NECESARIAS


```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
```

CARGAMOS LOS DATOS

```python
breast_cancer = load_breast_cancer()
X = breast_cancer.data
y = breast_cancer.target
breast_cancer.target_names
```

PREPROCESAMIENTO DE LOS DATOS Y PROGRAMAMOS UN PORCENTAJE DE 20% PARA LAS PRUEBAS PARA QUE EL 80% RESTANTE SEA PARA EL ENTRENAMIENTO

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=6)
```

SELECCIONAMOS Y ENTRENAMOS AL MODELO

```python
tree_model = DecisionTreeClassifier()
tree_model.fit(X_train, y_train)
```

COMPROBAMOS LA EFICIENCIA DEL ARBOL DE DECISIONES

```python
train_accuracy = tree_model.score(X_train, y_train)
print(f"Training Accuracy: {train_accuracy}")
```

PREPARAMOS LAS PREDICCIONES Y HACEMOS 3 PREDICCIONES PARA CORROBORAR LA EFICIENCIA

```python
y_pred = tree_model.predict(X_test)
```

```python
print(y_pred[1])
print(y_test[1])
print(y_pred[20])
print(y_test[20])
print(y_pred[-1])
print(y_test[-1])
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=breast_cancer.target_names))
```

VIZUALIZAMOS EL ARBOL DE DECISIONES

```python
plt.figure(figsize=(20, 15))
tree.plot_tree(tree_model, filled=True, feature_names=breast_cancer.feature_names, class_names=breast_cancer.target_names)  # Changed to tree_model, breast_cancer.feature_names, breast_cancer.target_names
plt.title("Decision Tree for Breast Cancer Prediction")
plt.savefig("decision_tree_breast_cancer.png", dpi=300)
plt.show()
```
